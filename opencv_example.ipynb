```python
#step1-;import required library
import cv2
#step1-;import required library
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow#for image display in colab
#step2:-upload image
from google.colab import files
uploaded=files.upload()

```

```python
#step3-read the upload image
filename=list(uploaded.keys())[0]
image=cv2.imread(filename) #opencv read in BGR format
cv2_imshow(image) #show the original image
```

```python
#step-4:convert to Grayscale
import cv2
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
#step5:-Resize the image to 300x300
a=[image,gray]
for i in a:
  resized = cv2.resize(i, (300, 300))
  cv2_imshow(resized)
```

```python
# Step 6: Draw Rectangle and Text
annotated = image.copy()
cv2.rectangle(annotated, (50, 50), (250, 250), (0, 255, 0), 3)
cv2.putText(annotated, 'Welcome', (60, 45), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
cv2_imshow(annotated)
```

```python
#step7:-save the modified image
cv2.imwrite('output_image.jpg',annotated)
print("Image saved as output_image.jpg")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
```

```python
#step3:upload image
uploaded=files.upload()
filename=list(uploaded.keys())[0]
```

```python
#step-4:load and prepare image
img=load_img(filename,target_size=[224,224])#resize the image
img_array=img_to_array
img_array=np.expand_dims(img_array,axis=0)#convert to batch of one
```

```python
#define the augumentation generator

datagen=ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.5,1.5],
    fill_mode='nearest'
)
```

```python
# Generate augmented images
i = 0
for batch in datagen.flow(img_array, batch_size=1):
    plt.figure(figsize=(2,2))
    plt.imshow(batch[0].astype('uint8'))
    plt.axis('off')
    i += 1
    if i > 5:
        break
```
